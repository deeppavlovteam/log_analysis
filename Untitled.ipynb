```python
from log_analyser.log_dataframe import LogDataFrame
from pathlib import Path
import re

template = r'^(\S+?)\s(\S+?)\s(\S+?)\s(\[.+?\])\s"GET\s(.+?)"\s(.+?)\s(.+?)\s(".*?")\s(".*?")\s(".+?")$'

ldf = LogDataFrame({'log_dataframe_columns':['a'], 'log_dir': '', 'pickle_file': '', 'hashes_file': ''})

access = sorted([p for p in Path('/home/ignatov/log_stuff/data/nginx/').resolve().glob('files-access.log*.gz')])
total = []

for file in access[-60:]:
    lines = ldf._read_file(file)
    parsed = re.findall(template, lines, flags=re.MULTILINE)
    total += parsed

requests = [request for ip_from, domain, _1, timestamp, request, response_code, bytes, ref, app, _2 in total]
```

```python
from urllib.parse import urlparse, parse_qs
def_dict = {}

urls = []
for request in requests:
    file_templ = r'^(.+?)[\s]'
    config_templ = r'.+?[\?\&]config=(.+?)[\s\&]'
    file_match = re.search(file_templ, request)
    file_name = file_match.group(1)
    urls.append(file_name)

for url in urls:
    x = urlparse(url)
    params = parse_qs(x.query)
    for key, val in params.items():
        value = def_dict.get(key, [])
        value += val
        def_dict[key] = value
```
